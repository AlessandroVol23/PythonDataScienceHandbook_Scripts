# Simple Scatter Plots


```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
import numpy as np
```

```python
# You can create scatter plots with plt.plot / ax.plot as well
x = np.linspace(0, 10, 30)
y = np.sin(x)
# You control it with the argument which represents the type of symbol used for plotting
plt.plot(x, y, "o", color="black");
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker'{0}''".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
# You can even combine scatter and line plots
plt.plot(x, y, "-ok");
```

```python
# Additional arguments to plt.plot have a wide range of properties
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=1)
plt.ylim(-1.2, 1.2);
```

## Scatter Plots with ``plt.scatter``

```python
plt.scatter(x, y, marker="o");
```

- Difference is that with ``plt.scatter`` you can set the properties individually mapped to the data


```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap="viridis")
plt.colorbar()
```

```python
# Example with iris data where each sample is one of three types of flowers
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2, s=100*features[3], c=iris.target, cmap="viridis")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
plt.colorbar();
```
