# In-Depth: Naive Bayes Classification

This chapter gives an In-Depth overview about the classification algorithm **Naive Bayes**.  
- Naive Bayes models are fast and simple algorithms
- Because they are fast, simple and have few tunable parameters you often use them for a quick-and-dirty baseline for classification problems

## Bayesian Classification

- The Bayes's theorem describes the relationship of conditional probabilities of stochastical quantities
- You want to find the probability of a label given some features
$$
P(L~|~{\rm features}) = \frac{P({\rm features}~|~L)P(L)}{P({\rm features})}
$$

If we want to decide between two labels ($L1$ and $L2$), one way of making this decision is to compute the ratio of the posterior probability of each label:

$$
\frac{P(L_1~|~{\rm features})}{P(L_2~|~{\rm features})} = \frac{P({\rm features}~|~L_1)}{P({\rm features}~|~L_2)}\frac{P(L_1)}{P(L_2)}
$$

We now create a naive model to compute $P({\rm features}~|~L_i)$ 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Gaussian Naive Bayes
One of the easiest classifier is Gaussian naive Bayes. The assumption is that data from each label is drawn from a simple Gaussian distribution

```python
from sklearn.datasets import make_blobs 
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="RdBu");
```

If you assume that the data is described by a Gaussian distribution with no covariance between the dimensions you just have to finding the mean and standard deviation. The result is shown in the following figure:

![(run code in Appendix to generate image)](../notebooks/figures/05.05-gaussian-NB.png)

The ellipses represent the Gaussian generative model for each label. This is a simple reciepe to compute the likelihood of 
$P(features~|L_1)$.  
  
This procedure is implemented in Scikit-Learn's ``sklearn.naive_bayes:GaussianNB`` estimator:

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y);
```

We generate some data and try to predict the labels

```python
rng = np.random.RandomState(0)
Xnew = [.6, -14] + [14, 18] * rng.rand(2000,2)
ynew = model.predict(Xnew)
```

```python
# Let's plot it
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="RdBu")
lim = plt.axis()
plt.scatter(Xnew[:,0], Xnew[:,1], c=ynew, s=20, cmap="RdBu", alpha=0.1)
plt.axis(lim);
```

Naturally, the Bayesian formalism allows probabilistic classification, which we can computer with using the ``preict_proba`` method:

```python
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```

The labels are giving the probabilities for the first and second level. 

## Multinomial Naive Bayes

- Multinomial naive bayes is another distribution to use on naive bayes
- The idea is the same as above, just this time the data distribution is with a best-fit multinomial distribution

### Example: Classifying Text
- Multinomial Naive Bayes is often used in text classification
- Features are often related to word counts or frequencies within the document

```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
```

```python
data.target_names
```

```python
categories =["talk.religion.misc", "soc.religion.christian", "sci.space", "comp.graphics"]

train = fetch_20newsgroups(subset="train", categories=categories)
test = fetch_20newsgroups(subset="test", categories=categories)
```

```python
print(train.data[5])
```

That we can use this data we have to convert each string into a vector of numbers. We will use the TF-IDF vectorizer and create a pipeline that attaches it to a multinomial naive Bayes classfier.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
# Vectorizinh (putting Strings into Vectors) -> Applying on Naive Bayes
model.fit(train.data, train.target)
labels = model.predict(test.data)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt="d", cbar=False, xticklabels=train.target_names, yticklabels=train.target_names)
plt.xlabel("true label")
plt.ylabel("predicted label");
```

The prediction between computer talk and space talk is already pretty good.   
The model still gets confused between talk about religion and talk about Christianity.  
  
  We now have the opportunity to use tools to determine the category for strings using the ``predict()`` method

```python
def predict_category(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category("sending a payload to the ISS")
```

```python
predict_category("discussing islam vs atheism")
```

```python
predict_category("determining the screen resolution")
```

```python
predict_category("don't know if using HDMI or VGA")
```

```python
predict_category("jesus")
```

## When to use Naive Bayes?

Since Naive Bayes is making strong assumptions about data, they will not perform as well as more complicated models. But it still has some advantages:
- Extremly fast for training and prediction
- Provide straightforward probalistic prediction
- Often very easy interpretable
- They have very few tunable parameters
  
  Often it is a good idea to start with a naive bayes algorithm as a baseline.
   
Naive Bayes performs well especially in following situations:
- When the naive assumptions actually mathe the data
- For very well seperated categories
- For very high dimensional data
