# Introduction Scikit-Learn

## How Scikit handles data
- Scikit handles data as tables

```python
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

- Each row refers to a single flower = One sample
- Number of rows = All flowers = ``n_samples``
- Each column = Feature
- Number of columns = ``n_features``

### Feature Matrix

- This table is like a two dimensional array or matrix
- Normally called ``X``
- The shape is normally ``[n_samples, n_features]``

### Target Array

- Target array is normally ``y``
- In PANDAS in a ``Series``
- The form is normally: ``[n_samples, n_targets]``
- The ``species`` column - the thing we want to predict with the features or measurements - is the column which would be considered as the target array

```python
%matplotlib inline
import seaborn as sns
sns.set()
sns.pairplot(iris, hue="species", size=1.5);
```

Now we extract the feature matrix and target array from the ``DataFrame`` and delete the species column

```python
# Removes the species column and puts the rest in X_iris
X_iris = iris.drop("species", axis=1)
X_iris.shape
```

```python
# y_iris are the labels to the features in X_iris
y_iris = iris["species"]
y_iris.shape
```

Now we have one table with:
- X_iris = 150 x 4 shape which is 150 Examples x 4 Features
- y_iris = 150 x 1 => which is 150 labels for the examples

## Scikit Learns Estimator API

Scikit has the following principles
- **Consistency**: All objects share a common interface drawn from a limited set of methods
- **Inspection:** All parameter values are exposed as public attributes
- **Limited object hierachy**: Only algorithms are represented as classes. Datasets are represented in standard formats (NumPy arrays, Pandas ``DataFrames``, SciPi sparse matrices) and parameter names use standard Strings
- **Composition**: Many ML tasks can be expressed as sequences of more fundamental algorithms. SciKit-Learn makes use of this wherever possible
- **Sensible defaults**: When models require user-specified parameters, the library defines an appropriate default value

Because of this SciKit Leanr is easy to use once the basic principles are understood

## Basics of the API

1. Choose a class of model by importing the appropriate estimator class from Scikit-Learn 
2. Choose model hyperparameters by instantiating this class with desired values
3. Arrange data into a features matrix and target vector following the discussion above
4. Fit the model to your data by calling the ``fit()`` method of the model instance
5. Apply the Model to new data:
  - Supervised: Predict labels for unkown data using the ``predict()`` method
  - Unsupervised: Transform or infer properties of the data using the ``transform()`` or ``predict()`` method

## Supervised Learning example: Simple Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

Let's walk through the steps:

### 1. Choose a class of model

  - Every class of a model is represented by a Python class
  - If we want to compute a simple linear regression we can import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

### 2. Choose model hyperparameters

Important: Class of model != Instance of model

When we chose a model we still can change some options like:
- Would we like to fit for the offset (i.e. y-intercept)?
- Would we like to model to be normalized?
- Would we like to preprocess our features to add model flexiblity?
- What degree of regularization would we like to use in our model?
- How many model components would we like to use?

These choices are called hyperparameters. 

For our linear regression example we can instantiate the ``LinearRegression`` class and specify that we will fit the intercept using the ``fit_intercept`` hyperparameter

```python
model = LinearRegression(fit_intercept=True)
model
```

### 3. Arrange data into a features matrix and target vector
- Target variable ``y`` is already in the correct form
- ``x`` needs the size ``[n_samples, n_features]``

```python
X = x[:, np.newaxis]
X.shape
```

### 4. Fit the model to your data
Fit the model with the ``fit()`` method

```python
model.fit(X, y)
```

- The ``fit()`` method computes some things internal
- The results are stored in model-specific attributes
- All attributes that were learned during ``fit()`` have trailing underscoes ``xxx_`` in Scikit learn

```python
# Slope = Steigung
model.coef_
```

```python
# Intercept = y-Schnittpunkt
model.intercept_
```

### 5. Predict labels for unkown data

- After training the model we want to evaluate the model on new data
- This can be done with the ``predict()`` method
- Our new data in this case is just a grid of x values

```python
xfit = np.linspace(-1, 11)
```

```python
# We need the data in the form [n_samples, n_features] again
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

Finally, lets visualize the results by plotting the raw data and the model fit

```python
plt.scatter(x,y);
plt.plot(xfit, yfit);
```

## Supervised Example: Iris classification

- This time we take the Iris data and try to answer the question:
**Given a model trained on a portion of the Iris data, how well can we predict the remaining labels?**

- We use an extremely simple generative model known as Gaussian naive Bayes, which proceeds by assuming each class is drawn from an axis-aligned Gaussian distribution
- It is very fast and has no hyperparameters to choose
  
    

- Furthermore we split the data into a training set and a testing set
- Scikit learn hast a ``train_test_split()`` function

```python
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)
```

```python
# Now we can follow our reciepe again
from sklearn.naive_bayes import GaussianNB   # 1. Choose model class
model = GaussianNB()                         # 2. Instantiate model
model.fit(Xtrain, ytrain)                    # 3. Fit model to data
y_model = model.predict(Xtest)               # 4. Predict on new data
```

We can use ``accuracy_score`` utility to see the fraction of predicted labels that match their true value

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest, y_model)
acc = acc*100
print("%.2f" % acc, "%")
```

## Unsupervised learning example: Iris dimensionality

- As an unsupervised example we take a look at reducing the dimensionality of the Iris data that is is easier to visulize
- Iris data is four dimensional -> Four features and one recorded for each sample

We use **PCA = Principal Component Analysis** here which is a fast linear dimensionality reduction technique

```python
# We follow our recipe again:
from sklearn.decomposition import PCA        # 1. Choose the model class
# n_components = Number of components to keep -> We want to have 2 dimensions
model = PCA(n_components=2)                 # 2. Instaniate the model with hyperparameters
model.fit(X_iris)                           # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(X_iris)              # 4. Transform the data to two dimensions
```

Now we will plot the results. We can insert the results into the original Iris ``DataFrame`` and use Seaborn's ``lmplot`` to show the results

```python
iris["PCA1"] = X_2D[:, 0]
iris["PCA2"] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue="species", data=iris, fit_reg=False);
```

## Unsupervised learning: Iris clustering
- We will use the **Gaussian micture model (GMM)**
- A GMM attemnpts to meld the data as a collection of Gaussian blobs

```python
# Recipe:
from sklearn.mixture import GMM      # 1. Choose the model class
model = GMM(n_components=3, 
            covariance_type="full")  # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                    # 3. Fit to data
y_gmm = model.predict(X_iris)        # 4. Determine cluster labels
```

```python
# Recipe:
from sklearn.mixture import GaussianMixture      # 1. Choose the model class
model = GaussianMixture(n_components=3, covariance_type="full")  # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                    # 3. Fit to data
y_gmm = model.predict(X_iris)        # 4. Determine cluster labels
```

We put it in a ``DataFrame`` again

```python
iris["cluster"] = y_gmm
sns.lmplot("PCA1", "PCA2", data=iris, hue="species", col="cluster", fit_reg=False);
```

## Example: Exploring Hand Written Digits

### Loading and Visualizing the data

```python
# Loading and visualizing the digits
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

The image data is a three-dimensional array: 1797 samples each consisting of 8 x 8 grid of pixels  
Let's visualize them:


```python
import matplotlib.pyplot as plz

fig, axes = plt.subplots(10, 10, figsize=(8,8),
                         subplot_kw={"xticks":[], "yticks":[]}, 
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap="binary", interpolation="nearest")
    ax.text(0.05, 0.05, str(digits.target[i]), transform=ax.transAxes, color="green")
```

- We need a two-dimensional arraay with form: ``[n_samples, n_features]``
- We treat each pixel in the image as its own features -> 8x8 = 64 features
- This is already built into the dataset in ``data`` and ``target``

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

- 1797 samples
- 64 features

### Dimensionality Reduction

- We'll reduce the dimensions to 2 to visulize the data
- We use **manifold learning algorithm** called **Isomap** 

```python
from sklearn.manifold import Isomap # 1. Choose model class
iso = Isomap(n_components=2)        # 2. Instantiate with parameters
iso.fit(digits.data)                # 3. Fit to data
data_projected = iso.transform(     # 4. Determine cluster labels?
    digits.data)
data_projected.shape
```

Now the data is two-dimensional. Let's plot the data to see if we can learn anything from it

```python
plt.scatter(data_projected[:, 0], data_projected[:, 1], c=digits.target,
           edgecolor="none", alpha=0.5, cmap=plt.cm.get_cmap("Spectral", 10))

plt.colorbar(label="digit label", ticks=range(10))
plt.clim(-0.5, 9.5);
```

### Classification on digits

- We will apply a classification algorithm on the digits
- We split the data into training and testing set -> Fit a Gaussian naive Bayes model

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

We predicted our model. Let's check the accurancy

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest, y_model)
print("%.2f" %acc, "%")
```
