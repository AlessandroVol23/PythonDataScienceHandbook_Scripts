# Hyperparameters and Model Validation

Basic recipe to apply a supervised machine learning model:
  
   
1. Choose a class of model
2. Choose model hyperparameters
3. Fit the model to the training data
4. Use the model to predict labels for new data
  
The first two parts are perhaps the most important part. 

## Model Validation
Model validation should check how good your model was. 

### Model validation in the wrong way
We start with a naive way for model validation

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

Next we choose the model and the hyperparameter. This time we will use a k-neighbors classifir with ``n_neighbors=1``.   
This model says "the label of an unkown point is the same as the label of its closest training point"

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
```

Now we traing the model, and use it to predict labels for data we already know

```python
model.fit(X, y)
y_model = model.predict(X)
```

Finally, we computer the fraction of correctly labeled points

```python
from sklearn.metrics import accuracy_score
accuracy_score(y, y_model)
```

Of course this measure is 100% because the model was trained **and** tested on the **same data**.  
That doesn't make sense. You have to test it on a different dataset

## Model validation the right way: Holdout sets

We split the data into test and training data wih ``train_test_split``

```python
from sklearn.cross_validation import train_test_split
# split the data with 50% in each set
X1, X2, y1, y2 = train_test_split(X, y, random_state=0, train_size=0.5)

# fit the model on one set of data
model.fit(X1, y1)

# evaluate the model on the second set of data
y2_model = model.predict(X2)
accuracy_score(y2, y2_model)
```

This time the accuracy is at about 90% which is still pretty good.   
This time it was tested on untrained data

### Problems with cross-validation

Since we are using about 50% of our data for validation we just use 50% to traing the model.   
One way to address this is to use cross validation, that is, to do a sequence of fits where each subset of the data is used both as a training set and as a validation set.   
So first you take X1 as training set and in the second trial you use it as a validation set

```python
# Here you can ee two validation trials
y2_model = model.fit(X1, y1).predict(X2)
y1_model = model.fit(X2, y2).predict(X1)
accuracy_score(y1, y1_model), accuracy_score(y2, y2_model)
```

We could take the mean here to have a more accurate score. This is called two-fold cross validation. You can even expand it to a five-fold cross validation. 

```python
from sklearn.cross_validation import cross_val_score
cross_val_score(model, X, y, cv=5)
```

## Selecting the best model

To answer the question: *If our estimator is underperforming, how should we move forward?*  
There are several possible answers:
- Use a more complicated / more flexibel model
- Use a less complicated / less flexibel model
- Gather more training samples
- Gather more data to add features to each sample  
  
It is often hard to know the answers to this questions. You have to evaluate the bis-variance trade-off

### The Bias-variance trade-off
  
You have to find the sweet spot between bias and variance. Here are two examples for high bias and high variance models. Both are pretty bad  

![](../notebooks/figures/05.03-bias-variance.png)

The left model tries to fit the data with a straight line which is almost impossible. -> Underfit of Data / High Bias  
  
The right model tries to fit it with a high-order polynomial function. The model has a high flexibility and hits every point with the line. -> Overfit tha data / High Variance

#### Validation vs. training score $R^2$ in High bias and High Variance


- High bias models have a pretty similar training and validation score
- High variance models have a big gap 
  
    
If we have some ability to tune the odel complexity, we would expect the training and validation score to behave as illustrated in the figure:
![](../notebooks/figures/05.03-validation-curve.png)

This is called the valudation curve, and it has following features:
- The training score is always higher than the validation score. 
- For very low model complexity (high-bias) the training data is underfit. Which means the model makes bad predictions
- For high model complexity (high-variance) the training data is overfit. The model predicts the **training** data very well but fails for unseen data
- For some intermediate value, the validation curve has a maximum. This level has the best trade-off between bias and variance

### Validation curves in Scikit-Learn
  
Now we will see an example with a polynomial regression model. Which is a generalized linear model in which the degree of the polynomial is a tunable parameter. A straight line for example is:
$$
y = ax + b
$$

A degree-3 polynomial fits a cubic curve to the data

$$
y = ax^3 + by^2 + cx + d
$$

In Scikit-Learn, we can implement this with a simple linear regression combined with the polynomial preprocesor. We will use a pipeline to string these operations together. 

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree), LinearRegression(**kwargs))
```

Now we create some data to which we will fit our model

```python
import numpy as np

def make_data(N, err=1.0, rseed=1):
    # random sample data
    rng = np.random.RandomState(rseed)
    X = rng.rand(N, 1) ** 2
    y = 10 - 1. / (X.ravel() + 0.1)
    if err > 0:
        y += err * rng.randn(N)
    return X, y

X, y = make_data(40)
```

Now we can visualize the data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn;
seaborn.set()

X_test = np.linspace(-0.1, 1.1, 500)[:, None]

plt.scatter(X.ravel(), y, color="black")
axis = plt.axis()
for degree in [1, 3, 5]:
    y_test = PolynomialRegression(degree).fit(X, y).predict(X_test)
    plt.plot(X_test.ravel(), y_test, label="degree={0},".format(degree))
    
plt.xlim(-0.1, 1.0)
plt.ylim(-2, 12)
plt.legend(loc="best");
```

You can see three different degrees in this chart. Now we have to find the a suitable trade-off between bias and variance.  
  
Therefore, we visualize the validation curve for this particular data and model. This can be done with the function ``validation_curve``, given a model, parameter name and a range to explore, this function will automatically compute both the training score and validation score across the range.

```python
from sklearn.learning_curve import validation_curve
degree = np.arange(0, 21)
train_score, val_score = validation_curve(PolynomialRegression(), X, y, "polynomialfeatures__degree", degree, cv=7)

plt.plot(degree, np.median(train_score, 1), color="blue", label="training score")
plt.plot(degree, np.median(val_score, 1), color="red", label="validation score")
plt.legend(loc="best")
plt.ylim(0, 1)
plt.xlabel("degree")
plt.ylabel("score");
```

Here we can see that the best fit is about degree 3

```python
plt.scatter(X.ravel(), y)
lim = plt.axis()
y_test = PolynomialRegression(3).fit(X, y).predict(X_test)
plt.plot(X_test.ravel(), y_test);
plt.axis(lim);
```

The solid lines show the new results. The fainter lines show the previous results with the smaller dataset.   
It is clear from the validation curve that the larger dataset can support much more complicated models. 
  
  
So we see that the behavior of the validation curve has two important inputs. The model complexity and the number of training points. It is always a good idea to explore the behavior of the model as a function fo the number of training points, which we can do by using increasingly larger subsets of the data to fit our model. A plot of the training / validation score with respect to the size of the training set is known as a learning curve.  
  
The behavior we expect from this curve is:
- A model of a given complexity will **overfit** a small dataset. The training score will be high while the validation score will be pretty low
- A model of a given complexity will underfit a large dataset. The training score will decrease, but the validation score will increase
- A model will never have a better validation score than training score. 
  
So normally we can expect that a learning curve looks like that:

![](../notebooks/figures/05.03-learning-curve.png)


Once the learning curve is convergence to a particular score more training data won't help you to improve the model

### Learning curves in Scikit-Learn


```python
from sklearn.learning_curve import learning_curve

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for i, degree in enumerate([2, 9]):
    N, train_lc, val_lc = learning_curve(PolynomialRegression(degree),
                                         X, y, cv=7,
                                         train_sizes=np.linspace(0.3, 1, 25))

    ax[i].plot(N, np.mean(train_lc, 1), color='blue', label='training score')
    ax[i].plot(N, np.mean(val_lc, 1), color='red', label='validation score')
    ax[i].hlines(np.mean([train_lc[-1], val_lc[-1]]), N[0], N[-1],
                 color='gray', linestyle='dashed')

    ax[i].set_ylim(0, 1)
    ax[i].set_xlim(N[0], N[-1])
    ax[i].set_xlabel('training size')
    ax[i].set_ylabel('score')
    ax[i].set_title('degree = {0}'.format(degree), size=14)
    ax[i].legend(loc='best')
```

With degree two it won't make sense to add more data.
  
You can just improve your model by making it more complex. Then you need to have more data

## Validation practice: Grid search

Normally you have more things to change when choosing a model and getting the accurancy. Here is an example of using grid search to find the optimal polynomial model. We will explore a 3D grid of model features:
- Polynomial degree
- Flag telling wheter to fit the intercept
- Flag telling wheter to normalize the problem

So we fit the data to the grid search, tell the model which parameter we want to have and the model tells us the best options

This can be set up using Scikit-Learn's ``GridSearchCV`` meta-estimator:

```python
from sklearn.grid_search import GridSearchCV
param_grid = {"polynomialfeatures__degree": np.arange(21),
             "linearregression__fit_intercept": [True, False],
             "linearregression__normalize": [True, False]}

grid = GridSearchCV(PolynomialRegression(), param_grid, cv=7)
```

Like a normal estimator we didn't applied to any data yet. We have to call the ``fit()`` method

```python
grid.fit(X, y);
```

Now we can ask for the best parameters as follows

```python
grid.best_params_
```

Now we can use the best model and show the fit to our data using code from before:

```python
model = grid.best_estimator_

plt.scatter(X.ravel(), y)
lim = plt.axis()
y_test = model.fit(X, y).predict(X_test)
plt.plot(X_test.ravel(), y_test, hold=True);
plt.axis(lim);
```
